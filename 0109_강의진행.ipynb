### (2) 복수조건 필터링

```python
condition1 = ratings['Occupation'] == 'Artist'
condition2 = ratings['Age'] >= 30
```

```python
# 논리연산자 사용시 에러 발생
# ratings[condition1 and condition2]
```

```python
# 비트연산자 &
# 직업이 Artist 이면서, 나이가 30살 이상인 경우만 선택해.
ratings[condition1 & condition2]
```

```python
condition1 & condition2
```

```python
# 비트 연산자 |
# or와 같은 연산을 한다.
# 행끼리 or 연산을 하여, 전체 행의 길이만큼 True/False
# 직업이 Artist 이거나, 나이가 30살 이상인 경우 "모두" 선택

ratings[condition1 | condition2]
```

```python
condition1 | condition2
```

```python
# loc 문법하에서, 행 필터링하는데 사용 가능
filtered_ratings = ratings.loc[condition1 & condition2,['Occupation','Age','Rating']]
```

```python
filtered_ratings.head()
```

```python
ratings.loc[ratings['Occupation'] == 'Artist', 'Genre'].value_counts()
```

```python
ratings.loc[ratings['Occupation'] == 'Artist','Rating'].mean()
```

```python
ratings.loc[ratings['Occupation'] == 'Doctor','Rating'].mean()
```

## [4] 함수로 값 변환하기

```python
# map 메서드 안에 들어 온 규칙대로, 값 변환

ratings['Satisfaction'] = ratings['Rating'].map( {1:'하',
                                                  2:'하',
                                                  3:'중',
                                                  4:'상',
                                                  5:'상'})
```

```python
ratings.head()
```

```python
ratings.drop(columns='Satisfaction',inplace=True)
```

```python
ratings.head()
```

```python
for idx, rating in enumerate(ratings['Rating']):
    if rating >= 4:
        ratings.loc[idx,'Satisfaction'] = '상'
    elif rating >= 3:
        ratings.loc[idx,'Satisfaction'] = '중'
    else:
        ratings.loc[idx,'Satisfaction'] = '하'
```

```python
ratings.head()
```
