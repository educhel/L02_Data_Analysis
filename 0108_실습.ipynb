## (1) 조건에 따른 데이터 확인

```python
# 유료 사용자(is_paid == True)의 나이만 출력해 보세요.

# 컨테이너 반복 -> users <list> 의 요소 <dict>를 하나씩 받아서,
# 조건 비교 -> <dict>의 특정 값 (is_paid) 비교
# 비교 만족시, 나이 출력 -> if 거친 후, 출력

for user in users:
    if user['is_paid'] == True:
    # if user['is_paid']:
        print(user['age'])
```

```python
# 나이가 30세 이상인 사용자의 수를 출력해 보세요.

# 조건 : age >= 30
cnt = 0

for user in users:
    if user['age'] >= 30:
        cnt += 1

print(cnt)
```

## (2) 간단한 요약 계산

> 기술통계
- 중심 경향치 중, 하나인 "평균" 계산

```python
# 전체 사용자의 평균 나이를 계산해 보세요.

total_age = 0

for user in users:
     total_age += user['age']

avg_age = total_age / len(users)
avg_age
```

```python
# 유료 사용자만의 평균 나이를 계산해 보세요.

# 조건이 하나 추가!
paid_users_age = []

for user in users:
    if user["is_paid"]:
        paid_users_age.append(user['age'])

paid_avg_age = sum(paid_users_age) / len(paid_users_age)
paid_avg_age
```
