### (5) embark_town이 Southampton이고, fare가 50 이상인 승객의 정보를 선택하세요.

```python
# 다중 선택

condition1 = titanic['embark_town'] == 'Southampton' # 비교연산
condition2 = titanic['fare'] >= 50 # 비교연산

titanic.loc[condition1 & condition2,:]
```

## [2] 필터링 실습

### (1) 생존한 승객 중(survived)에서 각 등급(pclass)별 승객 수를 구하세요.

```python
titanic['survived'].value_counts()
```

```python
titanic.loc[titanic['survived'] == 1,'pclass'].value_counts()
```

### (2) 타이타닉호의 승객 중에서 형제자매/배우자 (sibsp 참고)가 없고, 20세 이하인 승객의 평균 요금을 계산하세요.


```python
titanic.dtypes
```

```python
titanic.loc[(titanic['sibsp'] == 0)&(titanic['age']<=20),'fare'].mean()
```

```python
titanic.mean(numeric_only=True)
```

### (3) 생존한 승객 중 나이가 기록된 승객의 평균 나이를 구하세요.

```python
titanic.info()
```

```python
titanic.loc[titanic['survived'] == 1,'age'].mean()
```

### (4) 'Southampton'에서 탑승하고 생존한 승객 중, 남성과 여성의 비율을 구하세요.

```python
titanic.loc[
            (titanic['embark_town'] == 'Southampton') & (titanic['survived']==1), # 행 필터링
            'sex' # 열 선택
            ].value_counts(normalize=True) # 선택한 하나의 시리즈를 대상으로 값 세기 (비중)
```

```python
titanic.loc[
            (titanic['survived']==1), # 행 필터링
            'sex' # 열 선택
            ].value_counts(normalize=True) # 선택한 하나의 시리즈를 대상으로 값 세기 (비중)
```

```python
titanic['sex'].value_counts(normalize=True)
```
